For the NMR part

```python
import json

file_path = '/Users/davidsegura/git/Leveraging_LLMs_causality/cage_SAD_llm/precursor_combination_map_to_exp_data.json'

with open(file_path, 'r') as f:
    data_json = json.load(f)
```

```python
def get_combination_from_json(json_obj):
    precursor_combination = json_obj['precursor_combination']
    di_number = int(precursor_combination[1:])  
    tri_letter = precursor_combination[0]  
    return di_number, tri_letter
```

```python
experiment_combinations = []


for key, json_obj in data_json.items():
    precursor_combination = json_obj['precursor_combination']
    experiment_code = json_obj['experiment_code']
    

    tri_letter = precursor_combination[0]  
    di_number = precursor_combination[1:]  
 
    di_name = f"Di{di_number}"
    tri_name = f"Tri{tri_letter}"

    experiment_combinations.append({
        'experiment_code': experiment_code,
        'di_name': di_name,
        'tri_name': tri_name
    })

comb_df = pd.DataFrame(experiment_combinations)


```

```python
df_cage_366 = pd.merge(df_features_turbidity, comb_df, left_on=['experiment_number', 'di_name', 'tri_name'], right_on=['experiment_code', 'di_name', 'tri_name'], how='inner')
```

```python
df_cage_366['conversion_label'] = ''

for index, row in df_cage_366.iterrows():
    if ((df_conversion['di_name'] == row['di_name']) & (df_conversion['tri_name'] == row['tri_name'])).any():
        df_cage_366.at[index, 'conversion_label'] = 'partial'
    else:
        df_cage_366.at[index, 'conversion_label'] = 'complete'
```

```python
df_cage_366 = df_cage_366.drop('experiment_code', axis=1)
```

```python
df_cage_366 = df_cage_366.drop(['tri_conformer_id_3d', 'tri_conformer_rmsd_3d', 'di_conformer_id_3d', 'di_conformer_rmsd_3d'], axis=1)
```
