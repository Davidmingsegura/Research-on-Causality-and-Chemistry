```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

def create_venn_diagram(list1, list2, title="Venn Diagram"):

    set1 = set(list1)
    set2 = set(list2)
    common_elements = set1 & set2
    unique_to_set1 = set1 - common_elements
    unique_to_set2 = set2 - common_elements

   
    text_set1 = "\n".join(unique_to_set1)
    text_set2 = "\n".join(unique_to_set2)
    text_common = "\n".join(common_elements)


    plt.figure(figsize=(14, 10))
    venn = venn2(subsets=(len(unique_to_set1), len(unique_to_set2), len(common_elements)),
                 set_labels=('List 1', 'List 2'))
    
  
    venn.get_label_by_id('10').set_text(text_set1)
    venn.get_label_by_id('01').set_text(text_set2)
    venn.get_label_by_id('11').set_text(text_common)

  
    plt.title(title)
    plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt


data = [
    (1, False), (1, False), (1, False), (1, False), (1, False), (1, False), (1, True),
    (2, True), (2, True), (2, True), (2, True), (2, True), (2, False), (2, False), (2, False), (2, False), (2, False),
    (3, False), (3, True), (3, True)
]


df = pd.DataFrame(data, columns=["steps", "correctness"])

df['correctness'] = df['correctness'].astype(bool)

counts = df.groupby(["steps", "correctness"]).size().unstack(fill_value=0)


plt.figure(figsize=(10, 6))  
ax = counts.plot(kind='bar', stacked=False, color=["salmon", "lightgreen"], alpha=0.8)


plt.title("Number of Steps (Average) vs Correctness", fontsize=16)
plt.xlabel("Number of Reasoning Steps", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.legend(["Incorrect (False)", "Correct (True)"], fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=0)


for container in ax.containers:
    ax.bar_label(container, label_type='edge', fontsize=10, padding=3)


plt.show()
```
